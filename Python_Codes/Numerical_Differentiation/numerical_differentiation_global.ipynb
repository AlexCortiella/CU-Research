```python
####################### FUNCTIONS #######################

# Simulate dynamical system

def duffing(x, t, gamma=0.1, kappa=1, epsilon=5):
    return [
            x[1],
            -gamma * x[1] - kappa * x[0] - epsilon * x[0] ** 3
            ]

gamma=0.1
kappa=1
epsilon=5


def lorenz63(x, t, sigma=10, rho=28, beta=8/3):
    return [
        sigma * (x[1] - x[0]),
        x[0] * (rho - x[2]) - z[1],
        x[0] * x[1] - beta * x[2]
    ]

def finite_differences(t,x):
    
    #First order finite differences scheme
    
    x2 = x[1:]
    t2 = t[1:]
    
    x1 = x[0:-1]
    t1 = t[0:-1]
    
    dxdt = (x2 - x1)/(t2 - t1)
    t = t1
    
    return (t,dxdt)
```

```python
####################### GENERATE DATA #######################

# Simulate dynamics

t0, tf = 0, 10  # start and end
dt = 0.01  # time step
Nt = int(np.floor(tf-t0)/dt + 1) #Number of time instances
n = 2 #Number of state variables
t_span = np.linspace(t0, tf, Nt)
x0 = [0, 1]
X = odeint(duffing, x0, t_span)
x1 = X[:,0]
x2 = X[:,1]

#Compute true derivatives
dX = np.zeros((len(t_span),n))
dX[:,0] = x2
dX[:,1] =  -gamma * x2 - kappa * x1 - epsilon * x1 ** 3

#Corrupt states by adding noise --> Observation model y(t) = x(t) + e(t)
sigma = 0.1
E = sigma * np.random.randn(Nt,2)#Additive zero-mean white noise (Assumed Gaussian)

Y = X + E 
t = t_span
m = len(t)#Number of samples

y1 = Y[:,0]
y2 = Y[:,1]

#Plot dynamical system and observations
plt.figure(figsize = (6,6))
plt.plot(y1,y2,'r.')
plt.plot(x1[1:],x2[1:])
plt.title('Duffing oscillator')
plt.ylabel('x2')
plt.xlabel('x1')
plt.legend(['observations','dynamics'])
plt.show()
```

## Trend filtering algorithm

$\hat{\beta} = \text{argmin} \frac{1}{2} ||y - \beta||_2^2 + \lambda ||D^{(k+1)}\beta||_1$

```python
# TREND FILTERING ALGORITHM
matrix = importr('Matrix')
genlasso = importr('genlasso')

# Cubic smoothing spline regularized by lambda
r1_y = robjects.FloatVector(y1)
r1_t = robjects.FloatVector(t)

#Create trendfilter object with specific inputs
trendfilter = genlasso.trendfilter(y = r1_y, ord = 3, maxsteps = 5885, minlam = 0)
```

```python
#Evaluate fitted data to specific grid r1_t ( Need to convert into np.array )
lamb_lasso_path = np.array(robjects.r['predict'](trendfilter).rx2('lambda'))
n_lambdas = len(lamb_lasso_path)
y_tf = np.array(robjects.r['predict'](trendfilter).rx2('fit'))

#Compute derivatives via splines for all lambdas
dy_tf = np.zeros((m,n_lambdas))
error_dy_tf = np.zeros(n_lambdas)

for i, reg_lambda in enumerate(lamb_lasso_path):
    y_tf_ss = interpolate.splrep(t, y_tf[:,i], k=3, s=0)
    dy_tf[:,i] = interpolate.splev(t, y_tf_ss, der=1)
    error_dy_tf[i] = norm(dy_tf[:,i] - dX[:,0])/norm(dX[:,0])
    
#Find the optimal lambda
min_indx_tf = np.argmin(error_dy_tf)
```

```python
## Plot results
lambda_tf = lamb_lasso_path[min_indx_tf]

print("Solution computed for lambda = "+str(lambda_tf))
#Plot results
fig,axs = plt.subplots(2,1,figsize = (16,8))

axs[0].plot(t,dX[:,0])
axs[0].plot(t,dy_tf[:,min_indx_tf])
axs[0].set_title('Optimal solution')
axs[0].set_xlabel('t')
axs[0].set_ylabel('dx1/dt')
axs[0].legend(['True','Approximation'])

axs[1].loglog(lamb_lasso_path,error_dy_tf)
axs[1].set_title('Error vs. $\lambda$')
axs[1].set_xlabel('$\lambda$')
axs[1].set_ylabel('error')

```

## Tikhonov differentation

```python
## Plot results
lambda_td = lamb_tik_diff[min_indx_td]

print("Solution computed for lambda = "+str(lambda_td))
#Plot results
fig,axs = plt.subplots(2,1,figsize = (16,8))

axs[0].plot(t,dX[:,0])
axs[0].plot(t,dy_td[:,min_indx_td])
axs[0].set_title('Optimal solution')
axs[0].set_xlabel('t')
axs[0].set_ylabel('dx1/dt')
axs[0].legend(['True','Approximation'])

axs[1].loglog(lamb_tik_diff,error_dy_td)
axs[1].set_title('Error vs. $\lambda$')
axs[1].set_xlabel('$\lambda$')
axs[1].set_ylabel('error')
```

## Cubic smoothing spline

```python
max_lamb = np.max(lamb_lasso_path)
n_lambdas = 100
lamb_sspline = np.logspace(-10,8,n_lambdas)

# Cubic smoothing spline regularized by lambda
r1_y = robjects.FloatVector(y1)
r1_t = robjects.FloatVector(t)

#Compute derivatives via splines for all lambdas
dy_ss = np.zeros((m,n_lambdas))
y_ss = np.zeros((m,n_lambdas))

for i, reg_lambda in enumerate(lamb_sspline):  
    
    r_smooth_spline = robjects.r['smooth.spline'] #extract R function# run smoothing function

    kwargs = {"x": r1_t, "y": r1_y, "lambda": float(reg_lambda)}
    spline1 = r_smooth_spline(**kwargs)
    y_ss[:,i] = np.array(robjects.r['predict'](spline1,robjects.FloatVector(r1_t)).rx2('y'))
    
    y_ss_ss = interpolate.splrep(t, y_ss[:,i], k=3, s=0)
    dy_ss[:,i] = interpolate.splev(t, y_ss_ss, der=1)

#Compute error 
error_dy_ss = np.zeros(n_lambdas)

for i, reg_lambda in enumerate(lamb_sspline):
    error_dy_ss[i] = norm(dy_ss[:,i] - dX[:,0])/norm(dX[:,0])
    
#Find the optimal lambda
min_indx_ss = np.argmin(error_dy_ss)
```

```python
## Plot results
lambda_ss = lamb_sspline[min_indx_ss]

print("Solution computed for lambda = "+str(lambda_ss))
#Plot results
fig,axs = plt.subplots(2,1,figsize = (16,8))

axs[0].plot(t,dX[:,0])
axs[0].plot(t,dy_ss[:,min_indx_ss])
axs[0].set_title('Optimal solution')
axs[0].set_xlabel('t')
axs[0].set_ylabel('dx1/dt')
axs[0].legend(['True','Approximation'])

axs[1].loglog(lamb_sspline,error_dy_ss)
axs[1].set_title('Error vs. $\lambda$')
axs[1].set_xlabel('$\lambda$')
axs[1].set_ylabel('error')
    
```
